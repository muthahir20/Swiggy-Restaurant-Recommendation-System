```python
#rating column cleaning

df["rating"]= df["rating"].replace('--','NaN')

df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

mean_ratings = df.groupby('city')['rating'].transform('mean')

global_mean = df['rating'].mean()

df['rating'] = df['rating'].fillna(mean_ratings).fillna(global_mean)

df['rating'] = df['rating'].astype(float).round(1)
```

```python
#cost column cleaning

df["cost"] = df["cost"].str.replace("₹", "")

df['cost'] = pd.to_numeric(df['cost'], errors='coerce')

mean_costs = df.groupby('rating')['cost'].transform(lambda x: x.mean(skipna=True))

global_mean = df['cost'].mean(skipna=True)
df['cost'] = df['cost'].fillna(mean_costs).fillna(global_mean)

df['cost'] = df['cost'].astype(int)
```

```python
# Data encoding 
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder

# Read your CSV file
df = pd.read_csv("D:\python_vs\swiggy_recommendation\cleaned_data.csv")

# Split 'city' into 'sub_city' and 'main_city'
df[['sub_city', 'main_city']] = df['city'].str.split(',', n=1, expand=True)

# Handle missing values
df['sub_city'] = df['sub_city'].fillna('Unknown').str.strip()
df['main_city'] = df['main_city'].fillna(df['sub_city']).str.strip()

# One-hot encode 'main_city' and 'sub_city'
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False).set_output(transform="pandas")
encoded_cities = encoder.fit_transform(df[['main_city', 'sub_city']])

# Add encoded columns to original DataFrame
df = pd.concat([df, encoded_cities], axis=1)

# Drop the original 'city', 'main_city', and 'sub_city' columns
df.drop(columns=['city', 'main_city', 'sub_city'], inplace=True)
```

```python
import pandas as pd
from sklearn.preprocessing import MultiLabelBinarizer

# Load the City_encoded.csv file
df = pd.read_csv("D:\python_vs\swiggy_recommendation\City_encoded.csv")

# Check if 'cuisine' column exists
if 'cuisine' in df.columns:
    # Clean and split the cuisine column (splitting by commas)
    df['cuisine'] = df['cuisine'].str.split(',').apply(lambda x: [i.strip() for i in x])

    print("Cuisine column after splitting:")
    print(df[['cuisine']].head())  
    
    # Initialize the MultiLabelBinarizer
    mlb = MultiLabelBinarizer()

    # Apply the fit_transform to get the one-hot encoded columns
    cuisine_encoded = mlb.fit_transform(df['cuisine'])

    # Add the encoded cuisine columns back to the original dataframe
    df[mlb.classes_] = cuisine_encoded

    # Drop the original 'cuisine' column
    df.drop('cuisine', axis=1, inplace=True)

    # Display the final DataFrame
    print("DataFrame with cuisine columns encoded and original 'cuisine' dropped:")
    print(df.head())

    # Save the modified DataFrame to a new CSV
    df.to_csv("D:\python_vs\swiggy_recommendation\City_encoded_with_cuisine.csv", index=False)

else:
    print("The 'cuisine' column does not exist in the dataframe.")
```
